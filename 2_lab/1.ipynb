### Починаємо вивчати цикли
- for - ітераційний цикл - коли ми маємо задане число ементів та виконуємо над ними операції
- while - умовний цикл - виконання операції поки умова є істинною задача для вивчення циклів - є кубик, нам потрібно кинувши його з 3 спроб щоб - нам випало 4 є дві функції які дозволяють нам керувати виконанням циклів
- continue - дозволяє оминути одну з ітерацій циклу та перейти на початок виконання циклу
- break - викидує з циклю завершуючи його роботу

```python
from random import randint

for i in range(3):
    dice = randint(1, 6)
    print(f"Нам випало число {dice}")
    if dice == 4:
        print(f"Ми виграли, нам випало число {dice}")
        break
    print("Хай пощастить наступного разу!")
```

```python
i = -1
dice = None
while i <= 3 and dice != 4: # умова виконується на початку утерації
    dice = randint(1, 6)
    print(f"Нам випало число {dice}")
    i+=1

if dice == 4:
    print(f"Ми виграли, нам випало число {dice}")
else:
    print("Хай пощастить наступного разу!")
```

### Самостійне завдання за допомогою ChatGPT
Опис роботи:
1. Нова гра: починається новий раунд.
1. Підрахунок очок: за кожен раунд, коли гравець зібрав три однакові фрукти, він отримує 10 очок.
1. Запит на продовження: після завершення кожного раунду запитується, чи хоче гравець продовжити гру.
1. Вивід загальних очок: після виходу з гри виводяться загальні очки.


```python
from random import choice, shuffle

# Список фруктів
fruits = ["Яблуко", "Банан", "Апельсин", "Груша", "Ківі", "Персик"]
total_points = 0  # Загальний підрахунок очок

# Основний цикл гри
while True:
    basket = []  # Очищаємо корзину перед новим раундом
    print("\n--- Новий раунд! ---")
    
    # Генеруємо фрукти, поки не зберемо три однакові
    for _ in range(100):  # Максимум 100 спроб
        fruit = choice(fruits)  # Випадковий вибір фрукта
        basket.append(fruit)
        print(f"Ми зібрали: {fruit}")

        # Перевіряємо, чи є три однакові фрукти
        if basket.count(fruit) == 3:
            print(f"Вітаємо! Ви зібрали три {fruit}!")
            total_points += 10  # Додаємо 10 очок за раунд
            break
    else:
        print("На жаль, не вдалося зібрати три однакові фрукти.")

    # Перемішуємо фруктову корзину
    shuffle(basket)
    print("Корзина фруктів:", basket)

    # Виводимо по 5 фруктів за раз з корзини
    while basket:
        pack = basket[:5]  # Витягуємо 5 фруктів
        print(f"Ми взяли пакет фруктів: {pack}")
        basket = basket[5:]  # Видаляємо 5 фруктів з корзини

    # Питання про продовження гри
    continue_game = input("Чи хочете ви продовжити гру? (так/ні): ").strip().lower()
    if continue_game != 'так':
        break

print(f"Гра закінчена! Ваші загальні очки: {total_points}")
```

### Виловлювання помилок
- Пайтон це інтерпретаційна мова - кожна операція виконується одна за одною
- Початково, явні помилки будуть підсвічуватись але це нам не забороняє запустити код a = [1 2]
- ми хочумо запрограмувати ділення на наше випадкове число, і в нас може випадково виникати значення 0 - а на 0 ділити не можна

```python
from random import randint

# n = input("Введіть число")
n = 47
for i in range(96):
    try:
        k = float(n) / randint(-69, 69)
    except ZeroDivisionError:  # Обробка помилки
        print("Ділити на 0 не можна!")
        k = None
    
    print(f"{i}) Отримуємо значення: ", k)
```

```python
try:
    raise ValueError("Ми самі робимо помилку!")
    a = 1 + '1'
except (TypeError, ValueError):
    print("Так робити не можна!")
```

```python
er = [ValueError, TypeError, KeyError, NameError]
from random import choice

try:
    a = choice(er)
    raise a("Помилка")
except Exception as e:
    print(f"В цьому випадку в нас виникла помилка: {e}, {a}")
```

```python
try:
    a = 47 / 0
except ZeroDivisionError:
    a = None
finally:
    print(f"Результат ділення буде {a}")
```

### Розглядаємо функції та лямбди
- фунція є обєктом, який ми можемо передавати або викликати
- визначення функції завжди потребує окремого блоку який починається з def

```python
def fun_name(a:str) -> bool:
    # тут буде тіло функції
    print(f"hey {a}")
    return True

print(fun_name)
fun_name("hello")
```

```python
greetings = ["Привіт", "Хеллоу", "Здоров", "Доброго дня"]
def say_hello(a):
    return f"{a} всім!"

for n in greetings:
    print(say_hello(n))

l = [say_hello(i) for i in greetings]
print(l)

lam = lambda x: f"{x} всім!"
l2 = lambda x: x+2

print(lam, l2)
lam("Привіт")

l = [lam(i) for i in greetings]
print(l)
```

```python
print([(lambda x: f"{x} всім!")(i) for i in greetings])
```
